# Task 2:Build Advanced Preprocessing Function

Implement the following function:
def preprocess_text(text):
    pass

Required Features
Your function must include:
Remove numbers
Example: "I have 2 dogs" → "I have dogs"
Remove extra spaces
Example: "This is good" → "This is good"
Handle repeated characters
Example: "soooo goooood!!!" → "so good"
Convert text to lowercase
Example: "WOW!!! This is GREAT!!!" → "wow this is great"
Remove very short tokens
Remove words with length ≤ 2
Exception: keep meaningful words like "no", "not"
Remove URLs and email-like patterns
Example: "Visit http://example.com now" → "visit now"

```python
import re

def preprocess_text(text):
    """
    Cleans raw text and returns tokens + cleaned sentence
    """

    try:
        # Edge case handling
        if not isinstance(text, str) or text.strip() == "":
            return [], ""

        # Lowercase
        text = text.lower()

        # Remove URLs
        text = re.sub(r'http\S+|www\S+', '', text)

        # Remove emails
        text = re.sub(r'\S+@\S+', '', text)

        # Remove numbers
        text = re.sub(r'\d+', '', text)

        # Remove special characters & emojis
        text = re.sub(r'[^a-z\s]', '', text)

        # Handle repeated characters (soooo → so)
        text = re.sub(r'(.)\1{2,}', r'\1', text)

        # Remove extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        tokens = text.split()

        # Remove short words except 'no', 'not'
        tokens = [
            word for word in tokens 
            if len(word) > 2 or word in ['no', 'not']
        ]

        clean_sentence = " ".join(tokens)

        return tokens, clean_sentence

    except:
        return [], ""
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for s in test_sentences:
    tokens, clean_sentence = preprocess_text(s)
    
    print(f"""
Original Text: {s}
Cleaned Tokens: {tokens}
Cleaned Sentence: {clean_sentence}
""")
    
    results.append(tokens)
```

# Task 4: Token Analytics

For each processed sentence, compute:
Total number of tokens
Number of unique tokens
Average token length
Analysis Questions
Which sentence had the most noise?
Which sentence retained the most meaningful tokens after cleaning?

```python
import numpy as np

for i, tokens in enumerate(results):
    if len(tokens) == 0:
        continue

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = np.mean([len(word) for word in tokens])

    print(f"""
Sentence {i+1}
Total Tokens: {total}
Unique Tokens: {unique}
Average Token Length: {round(avg_len, 2)}
""")
```

# Task 6: Build Full Pipeline

def full_pipeline(text_list):
    pass

Expected Output
{
    "tokens": [...],
    "clean_sentences": [...]
}

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, sentence = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# Test pipeline
output = full_pipeline(test_sentences)
print(output)
```

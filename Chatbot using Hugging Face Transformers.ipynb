## Introduction

A chatbot is an AI-based system that can interact with users in natural language.  
In this project, we use a transformer-based model to generate human-like responses.

Transformers are deep learning models that process text using attention mechanisms and are widely used in Natural Language Processing (NLP).

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

tokenizer = AutoTokenizer.from_pretrained("microsoft/DialoGPT-medium")
model = AutoModelForCausalLM.from_pretrained("microsoft/DialoGPT-medium")
```

```python
print("Chatbot: Hello! I am your AI assistant. How can I help you today?")
```

## Model Used

We use **DialoGPT-medium**, a pre-trained conversational model developed by Microsoft.

- It is based on the GPT architecture  
- Trained on large-scale conversational data  
- Capable of generating dynamic and context-based responses  

```python
for step in range(5):  # loop for multiple interactions
    user_input = input("User: ")
    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Goodbye!")
        break

    # Encode input
    inputs = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors="pt")

    # Generate response
    outputs = model.generate(inputs, max_length=1000, pad_token_id=tokenizer.eos_token_id)

    # Decode response
    reply = tokenizer.decode(outputs[:, inputs.shape[-1]:][0], skip_special_tokens=True)
    print(f"Chatbot: {reply}")
```

## Working Pipeline

The chatbot follows this process:

User Input → Tokenization → Model Processing → Response Generation → Output Display

- User input is converted into tokens  
- The model processes the tokens  
- A response is generated  
- The response is decoded and displayed  

## Features

- Uses pre-trained transformer model  
- Generates dynamic responses (no hardcoding)  
- Interactive conversation with user  
- Exit condition using 'exit' or 'quit'
